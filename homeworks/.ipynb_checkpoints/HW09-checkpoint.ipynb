### Raw data

```python
import pandas as pd
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings = lambda *a, **kw: None

df = pd.read_csv('data/train.csv')

df.columns = ["label", "title", "lead"]
label_map = {1:"world", 2:"sport", 3:"business", 4:"sci/tech"}
def replace_label(x):
    return label_map[x]
df["label"] = df["label"].apply(replace_label) 
df["text"] = df["title"] + " " + df["lead"]
df.head()
```

## Data Visualization

Another common problem when working with text data is trying to get a sense of the structure of the data. In this context unsupervise methods can help.

**Why do you think this is the case?**
The data that we consider for these types of questions is usually very high-dimensional. However, in many cases, there might be quite some strucutre in the data that lies on a lower dimensional manifold. When we do a dimension reduction, we have the hope of uncovering this lower dimensional structure. This might then yield very informative insights.

```python
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

##TODO use tsne to create clusters of observations on a sample of 20000 news
##TODO plot the results

tsne = TSNE(n_components=2, n_iter=500, perplexity=25)
Xtsne = tsne.fit_transform(X.iloc[1:10000,:])
plot = sns.scatterplot(
    x=Xtsne[:,0], y=Xtsne[:,1],
    hue=y.iloc[1:10000],
    palette=sns.color_palette("hls", 4),
    legend="full",
    alpha=0.3
)

plt.legend(title='Label', loc='upper left', labels=["0", "1", "2", "3"])
plt.show(plot)

# Reduced the number of points to make sure that the code runs in reasonable time...

# You can see that the cluster property of t-SNE is not very clear in two dimensions. Maybe a better clustering can be seen
# in higher dimensions. The points are plotted fairly close to points from the same label, however, the 
# clusters have not really drifted apart, making it hard to distinguish them if we would not have had the labels!!
```

## Dimensionality Reduction

To provide some motivation for why we do dimensionality reduction, you can first try to run the following cell to train XGBoost with the whole set of fetures (I suggest you try but kill the code after 1 minute max)

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score

xgb = XGBClassifier()

Xtrain, Xtest, ytrain, ytest = train_test_split(X.iloc[1:5000,:], y.iloc[1:5000], test_size=0.2)
xgb.fit(Xtrain, ytrain)
print("Accuracy Score on Test Set: ", accuracy_score(xgb.predict(Xtest), ytest))
```

### Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2

##use the chi2 feature selection from sklearn to select 100 features

selector = SelectKBest(chi2, k=100)
X_filtered = selector.fit_transform(X ,y)
mask = selector.get_support(indices=True)
X_filtered = X.iloc[:,mask]
X.iloc[:,mask].head()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

xgb_fs = XGBClassifier(objective='multi:softmax')

##TODO train an xgboost model on the selected features to predict the labels
##TODO print the accuracy score on the test set



Xtrain, Xtest, ytrain, ytest = train_test_split(X_filtered, y, test_size=0.2)
xgb_fs.fit(Xtrain, ytrain)
print("Accuracy Score on Test Set: ", accuracy_score(xgb_fs.predict(Xtest), ytest))
```

### PCA

```python
from sklearn.decomposition import PCA
##compute the pca components that account for at least 50% of the variation the data

pca = PCA(n_components=.50)
Xpca = pca.fit_transform(X)
pca.n_components_
```

```python
xgb_pca = XGBClassifier(objective='multi:softmax')

##TODO train an xgboost model on the selected features to predict the labels
##TODO print the accuracy score on the test set

Xtrain, Xtest, ytrain, ytest = train_test_split(Xpca, y, test_size=0.2)
xgb_pca.fit(Xtrain, ytrain)
print("Accuracy Score on Test Set: ", accuracy_score(xgb_pca.predict(Xtest), ytest))
```

**Which method do you think works best?**

It seems that there is still much information in the features that are not selected. When we only train the model on the first 10000 data points, we get an accuracy of around 80%, while this drops to 71 percent for the chi2 feature selection. This, however, could also be beacuse the first 10000 data points are "easy" to predict.

Morevoer, we see that with retaining 50% of the variance with PCA, we can obtain an accuracy of about 80%. Hence PCA in this case seems to be better at summarizing the data in a lower dimensionality. IN addition, it only needs about 60 dimensions, while with the chi2 metric we took the 100 dimensions.

A reason for this could be that with the chi2 metric, we just throw away dimensions wihtout summarizing their information. PCA first finds a orthonogal basis for the complete dataset, and then we throw away low variance directions. Hence, we summarize the data more.
