The goal of this homework is to **predict** the value of houses in California starting from house and neighborhood caracteristics. 

Differently from last week's homework, we are not interested in investigating the causal effect of some features on house values, but the main purpose is to obtain good predictions of the values, so to train the best model to minimize the loss function. 

```python
# Load Data
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
print(data['DESCR'])
import matplotlib.pyplot as plt
```

```python
X_full, y_full = data.data, data.target
n_samples = X_full.shape[0]
n_features = X_full.shape[1]
print(n_samples,n_features)
```

# Summary Statistics

```python
from seaborn import distplot
# make a histogram of the label variable
#TODO
import seaborn as sns

fig1 = sns.histplot(data = y_full)
fig1.set(ylabel = "Count", xlabel = "Label")
```

```python
# make histograms of all predictors
log_scales = [False, False, False, False, False, True, False, False]
for i in range(X_full.shape[1]):
    print(data.feature_names[i])
    fig_i = sns.histplot(data = X_full[:,i], log_scale = log_scales[i])
    fig_i.set(ylabel = "Count", xlabel = data.feature_names[i])
    plt.show()
```

```python
from seaborn import scatterplot
import numpy as np
# make scatterplots of all predictors with the label

for i in range(X_full.shape[1]):
    print(data.feature_names[i])
    if log_scales[i] == True:
        fig_i = sns.scatterplot(x = np.log(X_full[:,i]), y = y_full)
    else:
        fig_i = sns.scatterplot(x = X_full[:,i], y = y_full)
    fig_i.set(xlabel = data.feature_names[i], ylabel = data.target_names)
    plt.show()
```

# Train/Test Split

```python
from sklearn.model_selection import train_test_split
# Split the data into 80% training and 20% test
#TODO

X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, test_size = 0.2, random_state = 123)
```

# Pre-Processing

```python
from sklearn.preprocessing import StandardScaler
# produce standardized predictors with mean zero and variance one 
# fit the transformer in the training set and apply to the whole data
#TODO

scaler = StandardScaler()

# Since we are using ridge regression, it makes sense to also de-mean y. Otherwise we have to fit an intercept and that is annoying with this function!
scaler_y = StandardScaler(with_std = False)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1))
y_test_scaled = scaler_y.transform(y_test.reshape(-1,1))
```

```python
from sklearn.preprocessing import PolynomialFeatures
# produce quadratic polynomial features from all standardized predictors
# squared values and pairwise interactions 
# fit the transformer in the training set and apply to the whole data
#TODO

transformer = PolynomialFeatures(degree = 2, include_bias = False) # If intercept is included here, it is penalized in ridge!
X_train_processed = transformer.fit_transform(X_train_scaled)
X_test_processed= transformer.transform(X_test_scaled)
```

## Fractional Ridge Regression
https://arxiv.org/abs/2005.03220

https://nrdg.github.io/fracridge/

```python
from fracridge import FracRidgeRegressorCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# train fractional ridge on raw predictors
# report the chosen fraction penalty
# report MSE and r-squared in test set 
#TODO

model_raw = FracRidgeRegressorCV()
model_raw.fit(X_train, y_train_scaled, frac_grid = [0.0001, 0.001, 0.01, 0.1, 0.4, 0.6, 0.8, 1])
print("MSE: ", mean_squared_error(y_test_scaled, model_raw.predict(X_test)))
print("R2:" , r2_score(y_test_scaled, model_raw.predict(X_test)))
print("Frac: ", model_raw.best_frac_)
```

```python
# train fractional ridge on stanardized predictors
# report the chosen fraction penalty
# report MSE and r-squared in test set 
#TODO

model_standardized = FracRidgeRegressorCV().fit(X_train_scaled, y_train_scaled, frac_grid = [0.0001, 0.001, 0.01, 0.1, 0.4, 0.6, 0.8, 1])
print("MSE: ", mean_squared_error(y_test_scaled, model_standardized.predict(X_test_scaled)))
print("R2:" , r2_score(y_test_scaled, model_standardized.predict(X_test_scaled)))
print("Frac: ", model_standardized.best_frac_)
```

```python
# train fractional ridge on standardized predictors including quadratics and interactions
# report the chosen fraction penalty
# report MSE and r-squared in test set 
#TODO

model_processed = FracRidgeRegressorCV().fit(X_train_processed, y_train_scaled, frac_grid = [0.0001, 0.001, 0.01, 0.1, 0.4, 0.6, 0.8, 1])
print("MSE: ", mean_squared_error(y_test_scaled, model_processed.predict(X_test_processed)))
print("R2:" , r2_score(y_test_scaled, model_processed.predict(X_test_processed)))
print("Frac: ", model_processed.best_frac_)
```
